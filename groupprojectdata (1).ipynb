```python
import numpy as np
```

```python
data = np.genfromtxt("groupdata.txt", 
                     dtype=[('day','i8'),('time','U6'),('height','f8')])

#print(data)
#print(data.shape)

#setting empty lists that I will fill up with the column data
day = []
time = []
height = []

for i in range(82):
    #assigning first column to the day variable
    day.append(data[i][0])
    #assigning second column to the time variable
    time.append(data[i][1])
    #assigning third column to the height variable
    height.append(data[i][2])
    
print(day)
print(' ')
print(time)
print(' ')
print(height)
```

```python
def convert_time(time_value):
    #seperating the time into hours and minutes based on the :
    hours, minutes = time_value.split(":")
    
    #transforming them from a string into a number
    hours = float(hours)
    minutes = float(minutes)
    
    #transform minutes into hours and add to hours
    hours = hours + minutes/60
    
    #return to the user the value of hours in days
    return hours/24

#creating an empty list for new xaxis
xaxis = []

for i in range(len(time)):
    time_converted = convert_time(time[i])
    
    #add each converted time to that same day element.
    xaxis.append(day[i] + time_converted)
```

```python
print(xaxis)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(xaxis,height, '-o')
plt.xlabel("Days")
plt.ylabel("Tide Height")
```
